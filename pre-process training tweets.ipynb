```python
import pandas as pd
```

```python
# Load the preprocessed dataset
encoding = 'ISO-8859-1'
train_df=pd.read_csv('/content/drive/MyDrive/tweets (1).csv', encoding=encoding)
```

```python
columns=['polarity','id','date','query','user','text'] 
train_df.columns=columns
train_df
```

```python
#Drop unwanted columns
columns=['id','date','query','user']
train_df= train_df.drop(columns,axis=1)
# replace the polarity values to 1 for positive and 0 for negative 
train_df['polarity'] = train_df['polarity'].replace({4: 1, 0: 0})
```

```python
#save the changes
train_df.to_csv('/content/drive/MyDrive/train.csv',index=False)
train_df
```
